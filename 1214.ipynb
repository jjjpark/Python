```python
def get_df(url):
    api_key = '466e4f4644636f773130306775434a63'
    url_re=url.replace('(인증키)',api_key).replace('xml','json').replace('/5','/1000')
    res = requests.get(url_re).json()
    df = pd.DataFrame(res[url.split('/')[-4]]['row'])
    return df
```

# 전월세 = 전세,월세 - 전세만 뽑아서 df 만들기
# astype -> 보증금을 astype으로 int,float 변경
# 분리된 df로 자치구명을 기준으로 보증금의 평균 구하기
# 보증금 높은 순으로 sort 정렬

```python
my_df.columns=['자치구명','법정동명','지번구분명','층수','계약일','전월세','면적','보증금','임대료','건물명','건축년도']
```

```python
my_df
```

```python
js = my_df[my_df['전월세'] == '전세']
```

```python
js
```

```python
js['보증금'] = my_df['보증금'].astype(int)
```

```python
js['임대료'] = my_df['임대료'].astype(int)
```

```python
my_df2 = js[['자치구명','보증금']]
```

```python
my_df2
```

```python
my_df2.groupby('자치구명').mean().round(0)
```

```python
my_df3 = my_df2.groupby('자치구명').mean().round(0)
```

```python
my_df3.sort_values(by = ['보증금'],ascending = False)
```
