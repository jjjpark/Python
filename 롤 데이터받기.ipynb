```python
def get_spell_data():
    # Riot Games API에서 스펠 데이터 받아오기
    url = "https://ddragon.leagueoflegends.com/cdn/14.2.1/data/ko_KR/summoner.json"
    response = requests.get(url)
    spell_data = response.json()["data"]

    return spell_data

def save_to_json(spell_data, file_path):
    # JSON 파일로 데이터 저장
    with open(file_path, 'w', encoding='utf-8') as json_file:
        json.dump(spell_data, json_file, ensure_ascii=False, indent=4)

def main():
    # 스펠 데이터 받아오기
    spell_data = get_spell_data()

    # JSON 파일로 저장
    save_to_json(spell_data, 'spell_data.json')
    print("Spell data saved to spell_data.json")

if __name__ == "__main__":
    main()

# x
```
