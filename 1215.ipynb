```python
df.columns = ['alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash', 'magnesium',
       'total_phenols', 'flavanoids', 'nonflavanoid_phenols',
       'proanthocyanins', 'color_intensity', 'hue',
       'od280_od315_of_diluted_wines', 'proline','drunk']
```

```python
def insert(x,conn):
    query = (
    f'insert into wine_table(alcohol,malic_acid,ash,alcalinity_of_ash,magnesium,total_phenols,flavanoids,nonflavanoid_phenols,'
    f'proanthocyanins, color_intensity, hue, od280_od315_of_diluted_wines, proline, drunk)'
    f'values({x.alcohol},{x.malic_acid},{x.ash},{x.alcalinity_of_ash},{x.magnesium},{x.total_phenols},{x.flavanoids},'
    f'{x.nonflavanoid_phenols},{x.proanthocyanins},{x.color_intensity},{x.hue},{x.od280_od315_of_diluted_wines},{x.proline},'
    f'{repr(x.drunk)}) '
    )
    mu.sql_execute(conn,query)
    return
```

```python
conn = mu.connect_mysql()
df.apply(lambda x:insert(x,conn) ,axis=1)
conn.commit()
conn.close()
```
